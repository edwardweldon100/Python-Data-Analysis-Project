```python
#Define career pay function (10 years till mid career pay)
#---assuming age 22 at start of career, 32 at mid-career, 65 at end of career.  (mid-career age can be adjusted)
#---assuming constant annual pay growth rate till end of career
def career_pay_w_4yr_degree(early_career_pay, mid_career_pay, length = '4 year', yrs_till_mid = 10):
    growth_rate = (mid_career_pay / early_career_pay) ** (1 / yrs_till_mid) - 1
    annual_pay = [early_career_pay]

    if length == '2 Year':
        for year in range(1, 45):
            year_pay = early_career_pay * (1 + growth_rate) ** year
            annual_pay.append(year_pay)
    else:
        for year in range(1, 43):
            year_pay = early_career_pay * (1 + growth_rate) ** year
            annual_pay.append(year_pay)
    
    return sum(annual_pay)
```

```python
#Define Present Value of career pay function (10 years till mid career pay)
#---assuming it's May of current year (graduation month)
#---assuming age 18 now, 22 at start of career, 32 at mid-career, 65 at end of career. (mid-career age can be adjusted)
#---assuming constant annual pay growth rate till end of career (applied at start of each year)
#---assuming paid bi-weekly (26 payments per year)
#---assuming 8% annual discount rate (real rate of return on S&P 500 over last 10 years)
def career_pay_w_4yr_degree_pv(early_career_pay, mid_career_pay, length = '4 year', yrs_till_mid = 10, discount_rate = 8):
    growth_rate = (mid_career_pay / early_career_pay) ** (1 / yrs_till_mid) - 1

    if length == '2 Year':
        annual_pay = [early_career_pay]
        for year in range(1, 45):
            year_pay = early_career_pay * (1 + growth_rate) ** year
            annual_pay.append(year_pay)
        
        annual_pay = np.array(annual_pay)
        biweekly_array = np.ones(26)
        biweekly_pay = []
        for i in annual_pay:
            biweekly_pmts = biweekly_array * (i/26)
            biweekly_pay.append(biweekly_pmts)
        biweekly_pay = np.array(biweekly_pay).reshape(-1)
        
        biweekly_pay_pv_multiplier = []
        for p in range((2*26),(47*26)):
            multiplier = 1/(((1+discount_rate/100)**(1/52))**p)
            biweekly_pay_pv_multiplier.append(multiplier)
    else:
        annual_pay = [early_career_pay]
        for year in range(1, 43):
            year_pay = early_career_pay * (1 + growth_rate) ** year
            annual_pay.append(year_pay)
        
        annual_pay = np.array(annual_pay)
        biweekly_array = np.ones(26)
        biweekly_pay = []
        for i in annual_pay:
            biweekly_pmts = biweekly_array * (i/26)
            biweekly_pay.append(biweekly_pmts)
        biweekly_pay = np.array(biweekly_pay).reshape(-1)
        
        biweekly_pay_pv_multiplier = []
        for p in range((4*26),(47*26)):
            multiplier = 1/(((1+discount_rate/100)**(1/52))**p)
            biweekly_pay_pv_multiplier.append(multiplier)
    
    career_pay_pv = np.multiply(biweekly_pay, biweekly_pay_pv_multiplier)

    career_pay_pv_total = 0
    for i in career_pay_pv:
        career_pay_pv_total = career_pay_pv_total + i
    return career_pay_pv_total
```

```python
#Define Total College Cost
def college_cost(length, annual_cost):
    if length == '2 Year':
        total_cost = 2 * annual_cost
    else:
        total_cost = 4 * annual_cost
    return total_cost
```

```python
#Define Present Value of annual college cost
#---assuming it's May of current year (graduation month)
#---assuming fall term payment due 4 months from May and and spring term payment due 4 months after fall payment
#---assuming annual tuition growth_rate of 4% and annual room and board growth rate of 4%
#---assuming 8% annual discount rate (real rate of return on S&P 500 over last 10 years)
def college_cost_pv(length, annual_cost, growth_rate = 4, discount_rate = 8):
    term_cost = annual_cost / 2
    growth_rate_num = growth_rate / 100
    if length == '2 Year':
        payments = [term_cost, term_cost, term_cost * (1 + growth_rate_num), term_cost * (1 + growth_rate_num)]
        month_till_payments_due = [4, 8, 16, 20]
        pmt_pv_multiplier = []
        for m in month_till_payments_due:
            multiplier = 1/(((1+discount_rate/100)**(1/12))**m)
            pmt_pv_multiplier.append(multiplier)
    else:
        total_cost = 4 * annual_cost
        payments_sub = [0,1,2,3]
        payment_years = [x for pair in zip(payments_sub, payments_sub) for x in pair]
        payments = []
        for i in payment_years:
            pmt = term_cost * ((1 + growth_rate_num) ** i)
            payments.append(pmt)
        month_till_payments_due = [4, 8, 16, 20, 28, 32, 40, 44]
        pmt_pv_multiplier = []
        for m in month_till_payments_due:
            multiplier = 1/(((1+discount_rate/100)**(1/12))**m)
            pmt_pv_multiplier.append(multiplier)
        
    return sum(np.multiply(payments, pmt_pv_multiplier))
```

```python
#Custom Functions for Box Plot and Smoothed out Histogram
def custom_boxplot(frame_name, col_name):

    # Horizontal box plot with annotations
    plt.figure(figsize=(12, 8))
     
    # Plotting the box plot with specified color and style
    sns.boxplot(x=frame_name[col_name], color='skyblue', showmeans=True, meanprops={"marker": "D", "markerfacecolor": "red",
                                                                                 "markeredgecolor": "red", "markersize":10})
     
    # Titles, labels, and legends
    plt.title(col_name, fontsize=16)
    plt.xlabel(col_name, fontsize=14)
    plt.yticks([])  # Hide y-axis tick labels
    plt.legend(handles=[Line2D([0], [0], marker='D', color='w', markerfacecolor='red', markersize=10, label='Mean')],
               loc='upper left', fontsize=14)
     
    plt.tight_layout()
    plt.show()

def custom_kde(frame_name, col_name):
    sns.kdeplot(frame_name[col_name], fill = True, label = col_name)
```

```python
#Define custom Heat Map function
def custom_heatmap(frame_name):
    # Calculate correlations
    correlation_matrix = frame_name.corr()
     
    # Set matplotlib figure
    plt.figure(figsize=(12, 8))
     
    # Generate Heat Map
    sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=.5, fmt=".2f", vmin=-1, vmax=1)
    plt.title(frame_name, fontsize=16)
    plt.show()
```
