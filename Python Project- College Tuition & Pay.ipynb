```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.lines import Line2D
plt.style.use('ggplot')
import seaborn as sns

Diversity_school = pd.read_csv('Diversity_school.csv')

def file_inspect(frame_name):
    #Glimpse, Blueprint, Statistical Snapshot
    print('Glimpse, Blueprint, Statistical Snapshot','\n')
    print(frame_name.head(),'\n')
    print(frame_name.info(),'\n')
    print(frame_name.describe(),'\n')
    
    #Null Values
    print('Null Values','\n')
    print(frame_name.isnull().sum(),'\n')
    
    #Shape
    print('Shape','\n')
    frame_name_rows, frame_name_columns = frame_name.shape
    print(f"The dataset comprises {frame_name_rows} properties described across {frame_name_columns} attributes.",'\n')
    
    #Data Types
    print('Data Types','\n')
    frame_name_data_types = frame_name.dtypes
    frame_name_type_counts = frame_name_data_types.value_counts()
    print(frame_name_data_types,'\n')
    print(frame_name_type_counts,'\n')
    
    #Data Dictionary
    print('Data Dictionary','\n')
    print(frame_name_type_counts)
    print(frame_name_data_types)

file_inspect(Diversity_school)

```

```python
Historical_tuition = pd.read_csv('Historical_tuition.csv')

file_inspect(Historical_tuition)
```

```python
Salary_potential = pd.read_csv('Salary_potential.csv')

file_inspect(Salary_potential)
```

```python
Tuition_cost = pd.read_csv('Tuition_cost.csv')

file_inspect(Tuition_cost)
```

```python
#Custom Functions for Box Plot and Smoothed out Histogram
def custom_boxplot(frame_name, col_name):

    # Horizontal box plot with annotations
    plt.figure(figsize=(12, 8))
     
    # Plotting the box plot with specified color and style
    sns.boxplot(x=frame_name[col_name], color='skyblue', showmeans=True, meanprops={"marker": "D", "markerfacecolor": "red",
                                                                                 "markeredgecolor": "red", "markersize":10})
     
    # Titles, labels, and legends
    plt.title(col_name, fontsize=16)
    plt.xlabel(col_name, fontsize=14)
    plt.yticks([])  # Hide y-axis tick labels
    plt.legend(handles=[Line2D([0], [0], marker='D', color='w', markerfacecolor='red', markersize=10, label='Mean')],
               loc='upper left', fontsize=14)
     
    plt.tight_layout()
    plt.show()

def custom_kde(frame_name, col_name):
    sns.kdeplot(frame_name[col_name], fill = True, label = col_name)
```

```python
#Smooth Histogram & Boxplot for In State Tuition
custom_kde(Tuition_cost,'in_state_tuition')
custom_boxplot(Tuition_cost,'in_state_tuition')
```

```python
#Smooth Histogram & Boxplot for Out of State State Tuition
custom_kde(Tuition_cost,'out_of_state_tuition')
custom_boxplot(Tuition_cost,'out_of_state_tuition')
```

```python
#Define custom Heat Map function
def custom_heatmap(frame_name):
    # Calculate correlations
    correlation_matrix = frame_name.corr()
     
    # Set matplotlib figure
    plt.figure(figsize=(12, 8))
     
    # Generate Heat Map
    sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=.5, fmt=".2f", vmin=-1, vmax=1)
    plt.title(frame_name, fontsize=16)
    plt.show()
```

```python
#Heat Map for Tuition_cost
custom_heatmap(Tuition_cost_table_numbers_only)
```

```python
#Heat Map for Tuition_cost
custom_heatmap(Salary_potential_table_numbers_only)
```
